```python
import pandas as pd

# Load the Climate Engine export
df = pd.read_csv('ClimateEngine.csv')

# The first column is Date, the rest are rainfall at 4 points
# Let's see the column names
print("Original columns:", df.columns.tolist())
print(f"Shape: {df.shape}")

# Rename for clarity
df.columns = ['date', 'point_1', 'point_2', 'point_3', 'point_4']

# Parse date
df['date'] = pd.to_datetime(df['date'])

# Since all 4 points are within ~50 meters, they fall in the same CHIRPS pixel.
# Best approach: average them for a single representative value
df['rainfall_mm'] = df[['point_1', 'point_2', 'point_3', 'point_4']].mean(axis=1)

# Create final format for notebook
output = pd.DataFrame({
    'plot_id': 'P001',  # Your main plot
    'date': df['date'],
    'rainfall_mm': df['rainfall_mm']
})

# Save
#output.to_csv('data/raw/chirps_rainfall.csv', index=False)
output.to_csv('chirps_rainfall.csv', index=False)

print("\nReformatted data:")
print(output.head(10))
print(f"\nTotal rows: {len(output)}")
print(f"Date range: {output['date'].min().date()} to {output['date'].max().date()}")
print(f"Rainfall stats (mm):")
print(output['rainfall_mm'].describe())
```
